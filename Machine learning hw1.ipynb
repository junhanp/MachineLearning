```python
from numpy import array
from scipy.linalg import svd
```

```python
#plt.rcParams['figure.figsize']=[16,8]
A = imread('image4.jpg')
#X = np.mean(A)

image = plt.imshow(A)
image.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = svd(A, full_matrices = False)
#print(U)
#print(S)
#print(VT)
```

```python
S = np.diag(S)
```

```python
j = 0
for r in (1, 5, 10, 50, 100):
    Xapp = U[:,:r] @ S[:r,:r] @ VT[:r,:]
    plt.figure(j+1)
    j+=1
    img = plt.imshow(Xapp)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r - ' +str(r))
    plt.show()
```
